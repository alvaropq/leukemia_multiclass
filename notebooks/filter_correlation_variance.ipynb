```python
import pandas as pd
```

```python
df = pd.read_csv('flat/GSE13159.csv')
```

```python
pip install feature-engine
```

```python
from feature_engine.selection import SmartCorrelatedSelection
```

```python
feature_selector = SmartCorrelatedSelection(
    variables=None,
    method="spearman",
    threshold = 0.5,
    missing_values = 'ignore',
    selection_method = "variance",
    estimator = None,
)
```

```python
X = df.drop('class',axis=1)
```

```python
Xcolunas = feature_selector.fit_transform(X)
```

```python
Xcolunas
```

```python
Xcolunas['class'] = df['class']
```
